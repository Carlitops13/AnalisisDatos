Generar una nueva columna con el producto de otras dos columnas.

Jugadores con buena puntería y mucho daño a jugadores

```python
fornite_df = pd.read_csv("fornite.csv")
```

```python
fornite_df['Accuracy'] = fornite_df['Accuracy'].str.replace('%', '').astype(float) / 100
```

```python
fornite_df[(fornite_df['Accuracy'] > 0.40) & (fornite_df['Damage to Players'] > 500)]
fornite_df.head()
```

Partidas con muchas eliminaciones pero poco material usado

```python
fornite_df[(fornite_df['Eliminations'] >= 3) & (fornite_df['Materials Used'] < 100)]
fornite_df.head()
```

Jugadores que recorrieron mucha distancia pero hicieron poco daño

```python
fornite_df[(fornite_df['Distance Traveled'] > 800) & (fornite_df['Damage to Players'] < 300)]
fornite_df.head()
```

Buen daño estructural con baja precisión

```python
fornite_df[(fornite_df['Accuracy'] < 25) & (fornite_df['Damage to Structures'] > 2000)]
fornite_df.head()
```

Jugadores que ayudaron mucho (assists o revives)

```python
fornite_df[(fornite_df['Assists'] >= 2) & (fornite_df['Revives'] >= 1)]
fornite_df.head()
```

 Partidas con pocos materiales recolectados pero mucho daño total

```python
fornite_df[(fornite_df['Materials Gathered'] < 100) & ((fornite_df['Damage to Players'] + fornite_df['Damage to Structures']) > 1000)]
fornite_df.head()
```

Jugadores con más de 30 hits y más de 10 headshots

```python
fornite_df[(fornite_df['Hits'] > 30) & (fornite_df['Head Shots'] > 10)]
fornite_df.head()
```

Partidas donde se recibió poco daño pero se infligió bastante

```python
fornite_df[(fornite_df['Damage Taken'] < 200) & (fornite_df['Damage to Players'] > 400)]
fornite_df.head()
```

Jugadores muy activos: gran distancia y alto uso de materiales

```python
fornite_df[(fornite_df['Distance Traveled'] > 700) & (fornite_df['Materials Used'] > 200)]
fornite_df.head()
```

Jugadores con precisión media pero alto rendimiento estructural

```python
fornite_df[(fornite_df['Accuracy'].between(25, 45)) & (fornite_df['Damage to Structures'] > 1500)]
fornite_df.head()
```
