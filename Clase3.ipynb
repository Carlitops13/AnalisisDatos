```python
sql_query = """
SELECT 
    Alumnos,
    (Programacion + "Análisis" + "Bases de Datos" + Etica) / 4.0 AS Promedio_General
FROM 
    notas
ORDER BY 
    Promedio_General DESC;
"""


try:
    df_promedios = pd.read_sql_query(sql_query, con)
    print("\nResumen de Promedios por Alumno:")
    print(df_promedios)
except Exception as e:
    print(f"\nOcurrió un error al ejecutar la consulta: {e}")
```
